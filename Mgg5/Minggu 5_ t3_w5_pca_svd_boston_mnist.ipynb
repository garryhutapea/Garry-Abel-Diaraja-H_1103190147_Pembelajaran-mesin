```python

#pca and svd
#check loss and accuracy
#max to 3 dim viz

import pandas as pd
import numpy as np

from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA, TruncatedSVD
from matplotlib import pyplot as plt
# mnist ?

from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)

df_x.describe
```

```python
df_x.head()
```

```python
df_y.head()
```

```python
#put into classifier and get the accuracy

reg = linear_model.LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y,test_size=0.2, random_state=4)
reg.fit(x_train, y_train)
```

```python
#check the accuracy
reg.score(x_test, y_test)
```

```python
#then dim rec using PCA
#from 18 col to 8 col

pca=PCA(n_components=8, whiten='True')
x=pca.fit(df_x).transform(df_x)
```

```python
df_x.head()
```

```python
pca.explained_variance_ 
```

```python
#put again after DIM REDUCTION into classifier and get the accuracy

reg = linear_model.LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x,df_y,test_size=0.2, random_state=4)
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test) #accuracy
```

```python
#do again with SVD 
svd = TruncatedSVD(n_components = 10)
x = svd.fit(df_x).transform(df_x)
reg = linear_model.LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x,df_y,test_size=0.2, random_state=4)
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test) #accuracy
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y,test_size=0.2, random_state=4)
rf = RandomForestClassifier(n_estimators=50)
rf.fit(x_train, y_train)
```

```python
#now lets reduce the dim from 800-an to 10 
#something wrong dari titik ini ke bawah (visualisasi), silakan dibetulkan

pca = PCA(n_components=10, whiten='True') 
x  = pca.fit(df_x).transform(df_x)

x_train, y_test, y_train, y_test = train_test_split(x,df_y,test_size=0.2, random_state=4)
rf = RandomForestClassifier(n_estimators=50)
rf.fit(x_train, y_train)
```

```python
# recalculate the accuracy for new dimension
pred = rf.predict(x_test)
s = y_test.values
count = 0
for i in range(len(pred)):
	if pred[i]==s[i]:
		count = count+1
```

```python
count/float(len(pred))
```

```python
#Prepare to viz MNIST with PCA in 2 dimensions
pca=PCA(n_components=2, whiten='True')
x=pca.fit(df_x).transform(df_x)

x_train, y_test, y_train, y_test = train_test_split(x,df_y,test_size=0.2, random_state=4)
rf = RandomForestClassifier(n_estimators=50)
rf.fit(x_train, y_train)

pred = rf.predict(x_test)
s = y.test.values
count = 0
for i in range(len(pred)):
	if pred[i]==s[i]:
		count = count+1

```

```python
#check the accuracy first
count/float(len(pred))
```
